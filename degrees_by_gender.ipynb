# degrees by gender

### the file

```python
file = "data"
```

### importing the modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### file in dataframe

```python
women_degrees = pd.read_csv(file)
```

### majors represented

```python
for idx, major in enumerate(women_degrees.columns[1:]): 
    print ("{}: {}".format(idx+1, major))
```

### first few rows

```python
women_degrees.head(3)
```

```python
# let us create figure
fig = plt.figure(figsize=(20,24))

# let us create three lists of the majors
stem_majs = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
libArts_majs = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_majs = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# we will loop over these putting each its own column

# let us start the first loop
for idx, i in enumerate(range(0,18,3)):
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_majs[idx]], linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_majs[idx]], linewidth=3)
    ax.set_title(stem_majs[idx])

# going onto the second loop for libArts_majors
for idx, i in enumerate(range(1,14,3)):
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[libArts_majs[idx]], linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[libArts_majs[idx]], linewidth=3)
    ax.set_title(libArts_majs[idx])
    
# going onto the third loop of other_majors
for idx, i in enumerate(range(2,18,3)):
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[other_majs[idx]], linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_majs[idx]], linewidth=3)
    ax.set_title(other_majs[idx])
```
